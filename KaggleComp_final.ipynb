```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Loading the data
df = pd.read_csv("train.csv")
df.head()
#df.shape
```

```python
null_columns=df.columns[df.isnull().any()]
print(df[df.isnull().any(axis=1)][null_columns].head())
#NaN values are in three columns (purchase_date,total_positive_reviews,total_negative_reviews)
```

```python
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
df['release_date'] = pd.to_datetime(df['release_date'])
df
df['day']=df['purchase_date']-df['release_date']   #Find duration between the purchase_date and release date
df['day'] = df['day'] / np.timedelta64(1, 'D')
df
```

```python
df['day']=df['day'].fillna(df['day'].mean())
df['total_positive_reviews']=df['total_positive_reviews'].fillna(df['total_positive_reviews'].mean())
df['total_negative_reviews']=df['total_negative_reviews'].fillna(df['total_negative_reviews'].mean()) 
```

```python
df=df.drop(['purchase_date','release_date'],axis=1)
```

```python
col = ['is_free','genres','categories','tags']
df.loc[:,col]=df.loc[:,col].apply(lambda x: x.astype(str).str.upper())
```

```python
df['is_free'] =df['is_free'].apply(lambda x: 1 if x=='TRUE' else 0)
```

```python
cols = ['genres','categories','tags']
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation,' ')
    return text
for i in cols:
    df.loc[:,i]=df.loc[:,i].apply(remove_punctuations)
df.head()
```

```python
vect=CountVectorizer()
trans=vect.fit_transform(df['genres'])
vect.get_feature_names()
df2=pd.DataFrame(trans.toarray(),columns=vect.get_feature_names())

vect=CountVectorizer()
trans1=vect.fit_transform(df['categories'])
vect.get_feature_names()
df3=pd.DataFrame(trans1.toarray(),columns=vect.get_feature_names())
df3
vect=CountVectorizer()
trans2=vect.fit_transform(df['tags'])
vect.get_feature_names()
df4=pd.DataFrame(trans2.toarray(),columns=vect.get_feature_names())
df4
df_new = pd.concat([df,df2,df3,df4],axis=1)
df_new
df1=df_new.drop(columns=['id','genres','categories','tags'])
df1
df1_col=df1.columns
df1
```

```python

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.linear_model import PassiveAggressiveRegressor
```

```python
#Train model
model=PassiveAggressiveRegressor()
X_train=df1.iloc[:,1:442]
y_train=df1['playtime_forever']

model.fit(X_train, y_train)
```

```python
dff = pd.read_csv("test.csv")
dff.head()
null_columns=dff.columns[dff.isnull().any()]
dff['purchase_date'] = pd.to_datetime(dff['purchase_date'])
dff['release_date'] = pd.to_datetime(dff['release_date'])
dff['day']=dff['purchase_date']-dff['release_date']   #Find duration between the purchase_date and release date
dff['day'] = dff['day'] / np.timedelta64(1, 'D')
dff['day']=dff['day'].fillna(dff['day'].mean())
dff['total_positive_reviews']=dff['total_positive_reviews'].fillna(dff['total_positive_reviews'].mean())
dff['total_negative_reviews']=dff['total_negative_reviews'].fillna(dff['total_negative_reviews'].mean())
dff=dff.drop(['purchase_date','release_date'],axis=1)
col = ['is_free','genres','categories','tags']
dff.loc[:,col]=dff.loc[:,col].apply(lambda x: x.astype(str).str.upper())
dff['is_free'] =dff['is_free'].apply(lambda x: 1 if x=='TRUE' else 0)
cols = ['genres','categories','tags']
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation,' ')
    return text
for i in cols:
    dff.loc[:,i]=dff.loc[:,i].apply(remove_punctuations)
dff
vect=CountVectorizer()
form=vect.fit_transform(dff['genres'])
vect.get_feature_names()
df5=pd.DataFrame(form.toarray(),columns=vect.get_feature_names())

vect=CountVectorizer()
form1=vect.fit_transform(dff['categories'])
vect.get_feature_names()
df6=pd.DataFrame(form1.toarray(),columns=vect.get_feature_names())
df6
vect=CountVectorizer()
form2=vect.fit_transform(dff['tags'])
vect.get_feature_names()
df7=pd.DataFrame(form2.toarray(),columns=vect.get_feature_names())
df7
df_new2 = pd.concat([dff,df5,df6,df7],axis=1)
df_new2
dff2=df_new2.drop(columns=['id','genres','categories','tags'])
dff2
print(dff2)
#Columns that exists in training but not in testing
list=[]
for c in df1_col:
    if c not in dff2:
        list.append(c)
print(list)
print(len(list))

#Columns that exists in testing but not in training
list2=[]
for n in dff2:
    if n not in df1_col:
        list2.append(n)
print(list2)
list.remove('playtime_forever') 
print(list)
a = np.zeros(shape=(90,115))
dfl= pd.DataFrame(a,columns=list)
dfl.astype(int)
```
